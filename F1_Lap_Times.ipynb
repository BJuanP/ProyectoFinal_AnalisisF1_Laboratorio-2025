```python
# Renombrando las columnas de df_lap_times

df_lap_times.rename(columns ={
    'raceId': 'id_carrera',
    'driverId': 'id_piloto',
    'lap': 'vuelta',
    'position': 'posicion',
    'time': 'tiempo',
    'milliseconds': 'milisegundos'
}, inplace=True)


print("Transformación de datos exitosa")
```

```python
# Agregando la columna de tiempos de vuelta en segundos al df

def convertir_a_segundos(t):
    try:
        minutos, segundos = t.split(':')
        return int(minutos) * 60 + float(segundos)
    except:
        return None  # Maneja valores vacíos o mal formateados

df_lap_times['tiempo_segundos'] = df_lap_times['tiempo'].apply(convertir_a_segundos)


print("Transformación de datos exitosa")
```

```python
# Cambiando el tipo de dato de tiempo a string
# Voy a usar como guía visual el valor de tiempo de vuelta y para tratamiento usaré el tiempo en segundos

df_lap_times['tiempo'] = df_lap_times['tiempo'].astype('string')


print("Transformación de datos exitosa")
```

## 1 - ¿Cuales son las circuitos con mayor cantidad de abandonos?

```python
# Calcular la cantidad máxima de vueltas por carrera
vueltas_totales = df.groupby(['nombre_gp', 'fecha_carrera', 'ubi_circuito'], as_index=False)['vuelta'].max()
vueltas_totales.rename(columns={'vuelta': 'vueltas_totales'}, inplace=True)


# Calcular cuántas vueltas completó cada piloto en cada carrera
vueltas_pilotos = df.groupby(['nombre_gp', 'fecha_carrera', 'nombre_piloto'], as_index=False)['vuelta'].max()
vueltas_pilotos.rename(columns={'vuelta': 'vueltas_completadas'}, inplace=True)


# Unir ambos para comparar
df_abandonos = vueltas_pilotos.merge(vueltas_totales, on=['nombre_gp', 'fecha_carrera'], how='left')


# Obtener los abandonos (si no completó todas las vueltas)
df_abandonos['abandono'] = df_abandonos['vueltas_completadas'] < df_abandonos['vueltas_totales']


# Contabilizar abandonos por circuito
aband_por_circuito = df_abandonos[df_abandonos['abandono'] == True].groupby(['nombre_gp', 'ubi_circuito'], as_index=False)['abandono'].count()
aband_por_circuito.rename(columns={'abandono': 'cantidad_abandonos'}, inplace=True)


# Ordenamiento
aband_por_circuito.sort_values('cantidad_abandonos', ascending=False, inplace=True)


print("10 Circuitos con más abandonos")
aband_por_circuito.head(10)
```

## Conclusión


### Interpretación

En base a los resultados obtenidos puedo determinar que el **circuito con mayor cantidad de abandonos** es el del **Gran Premio de Austria**.  
También aparecen circuitos como los de **España**, **Bahréin** y **Abu Dhabi**, donde las condiciones climáticas extremas y el **mayor desgaste del asfalto** influyen directamente en el rendimiento.

Por otro lado, circuitos como los de **Hungría** o **Mónaco** seguramente se encuentran en el ranking debido a la **dificultad para adelantar**, lo que aumenta la probabilidad de **choques o contacto entre vehículos**.

---

### Propuesta

Como posible respuesta o medida de mejora se podría considerar **reducir el tamaño de los coches actuales**, ya que con el tiempo se hicieron cada vez más grandes y esto dificulta las maniobras, lo que aumenta el riesgo de incidentes en pistas estrechas.


## 2 - ¿Cuáles son los circuitos con mayor variabilidad en los tiempos por vuelta? (desviación del tiempo promedio de las vueltas)

```python
# Calcular promedio y desviación estándar de tiempo por vuelta en cada circuito
tiempos_por_circuito = (
    df.groupby('ubi_circuito', as_index=False)
    .agg(
        promedio_tiempo=('tiempo_segundos', 'mean'),
        desviacion_tiempo=('tiempo_segundos', 'std')
    )
)


# Redondear los valores
tiempos_por_circuito['promedio_tiempo'] = tiempos_por_circuito['promedio_tiempo'].round(3)
tiempos_por_circuito['desviacion_tiempo'] = tiempos_por_circuito['desviacion_tiempo'].round(3)


# Ordenar por mayor desviación
tiempos_por_circuito.sort_values('desviacion_tiempo', ascending=False, inplace=True)


# Mostrar por pantalla
print("Circuitos con mayor variabilidad en tiempos por vuelta")
tiempos_por_circuito.head(10)
```

## Conclusión

### Interpretación

En el primer lugar, y llegando casi a duplicar al segundo, está el **circuito de Mugello** (Gran Premio de la Toscana).  
Solo se disputó **una carrera en este circuito** (año 2020), y fue reconocida por ser **altamente caótica**, con múltiples accidentes y periodos de bandera roja, lo que podría explicar la gran dispersión en los tiempos de vuelta.

Circuitos como **Silverstone** o **São Paulo** también están presentes, ya que son caracterizados por su **clima impredecible** y a la **frecuencia de interrupciones** (safety car o incidentes menores).

---

### Propuesta

Para reducir la inestabilidad en los tiempos de carrera se podría tener en cuenta la configuración de estos circuitos o del asfalto de los mismos. También se podría plantear cambiar las fechas del calendario de cierto modo de que se minimice la probabilidad de lluvias o cambios en el clima.


## 3 - ¿Cual es la nacionalidad más predominante de los pilotos en la F1?

```python
# Contamos pilotos únicos por nacionalidad en ese rango temporal
nacionalidades = (
    df.groupby('nacionalidad')['nombre_piloto']
    .nunique()
    .reset_index()
    .rename(columns={'nombre_piloto': 'cantidad_pilotos'})
    .sort_values('cantidad_pilotos', ascending=False)
)


# Mostrar por pantalla
print("Nacionalidades con más predominantes de los pilotos")
nacionalidades.head(5)
```

## Conclusión

### Interpretación

La actual región más predominante en la F1 es el Reino Unido, con un total de 9, seguida de varias regiones como Alemania (5), Brasil (4) y Francia (4)

Países como el Reino Unido destacan por tener una academias de pilotos y programas juveniles, por lo que salen mucho mejor preparados que en otras regiones donde no se cuenta con los recursos.
